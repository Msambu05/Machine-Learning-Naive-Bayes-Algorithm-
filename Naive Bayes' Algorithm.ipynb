**Determine whether the new object with features (color:Red, Size:Small, Texture:Rough, Weight:Light) is an apple. Show all the steps that made you to arrive to the final answer.**
# Step 1: Calculate Prior probability

    P(Apple = Yes) = Number of Yes in Apple / Total instances
    		= 5/10
    		= 0.5
    
    P(Apple = No) = Number of No in Apple / Total instances
    		= 5/10
    		= 0.5


# Step 2: Calculate Likelihood probability

**2.1. For Yes Probability of the target(Apple)**

	P(Color = Red | Apple = Yes) = (Color=Red and Apple=Yes) / (Apple=yes)
					= 5/5
					= 1
	
	P(Size = small | Apple = yes) = (Size=Small and Apple=Yes) / (Apple=Yes)
					= 3/5
					= 0.6

	P(Texture = Rough | Apple = Yes) = (Texture=Rough and Apple=Yes) /(Apple=Yes)
					         = 2/5
					         = 0.4

	P(Weight = Light | Apple = Yes) = (Weight=Light and Apple=Yes) / (Apple=Yes)
					     = 2/5
					     = 0.4


**2.2. For No Probability of the target(Apple)**

	P(Color = Red Apple = No) = (Color=Red and Apple=Yes) / (Apple=No)
					= 0/5
					= 0
	
	P(Size = small | Apple = No) = (Size=Small and Apple=Yes) / (Apple=No)
					= 2/5
					= 0.4

	P(Texture = Rough | Apple = No) = (Texture=Rough and Apple=Yes) /(Apple=No)
					         = 3/5
					         = 0.6

	P(Weight = Light | Apple = No) = (Weight=Light and Apple=Yes) / (Apple=No)
					     = 3/5
					     = 0.6
	
 	

# Step 3: Calculate Posterior probability 

**3.1. For Yes Probability on the target**

	= 1 * 0.6 * 0.4 * 0.4 * 0.5
	= 0.048

**3.2. For No Probability on the target**
	
	= 0 * 0.4 * 0.6 * 0.6 * 0.5
	= 0
	
# Step 4: Normalize Posterior probability

    P(Apple = Yes) = P(Apple=Yes) / color=Red,Size=Small,Texture=Rough,Weight=Light
	               = 0.048 / 0.048 + 0			 
	               = 1

    P(Apple = No) = P(Apple=No) / color=Red,Size=Small,Texture=Rough,Weight=Light
		          = 0 / 0 + 0.048
		          = 0

# Step 5: Provide the probability answer
	
   The probability of the new object being an apple having features including color = red, size = small, texture = rough, and weight = light, is 1         whereas the probability of the new object not being an apple having the exact features is 0, therefore the new object is an apple since 1 > 0.
